```python
import json as js
import numpy as np
import pandas as pd
import random as rd
files = ["A.fna", "B.fna"]
DNA = {}
for f in files:
  with open(f, 'r') as fp:
    seq = "".join(fp.read().splitlines())
  DNA[f] = {"seq" : seq}
  DNA[f]["char"] = {}
  for c in set(seq):
    DNA[f]["char"][c] = seq.count(c)
  index = list(DNA[f]["char"].keys())
  DNA[f]["t_mtx"] = {}
  for c in index:
    DNA[f]["t_mtx"][c] = {}
    for d in index:
      DNA[f]["t_mtx"][c][d] = 0
  for i, c in enumerate(seq[:-1]):
    DNA[f]["t_mtx"][c][seq[i + 1]] += 1
  for i in DNA[f]["t_mtx"].keys():
    s = np.sum(list(DNA[f]["t_mtx"][i].values()))
    for j, k in DNA[f]["t_mtx"][i].items():
      DNA[f]["t_mtx"][i][j] /= s
  print(js.dumps(DNA[f]["t_mtx"], indent=2))
  t_mtx = pd.DataFrame(DNA[f]["t_mtx"], index=index, columns=index)\
            .sort_index().sort_index(axis=1)
  print(t_mtx)

for f in files:
  seq = list(DNA[f]["seq"])
  rd.shuffle(seq)
  seq = "".join(seq)
  t_mtx = {}
  for c in index:
    t_mtx[c] = {}
    for d in index:
      t_mtx[c][d] = 0
  for i, c in enumerate(seq[:-1]):
    t_mtx[c][seq[i + 1]] += 1
  for i in t_mtx.keys():
    s = np.sum(list(t_mtx[i].values()))
    for j, k in t_mtx[i].items():
      t_mtx[i][j] /= s
  t_mtx = pd.DataFrame(t_mtx, index=index, columns=index).sort_index()\
            .sort_index(axis=1)
  print(t_mtx)
```
